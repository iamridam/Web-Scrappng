```python
import requests 
from bs4 import BeautifulSoup 
```

```python
URL = "https://www.drugscontrol.org/banned-drugs.php"
r = requests.get(URL) 
```

```python
soup = BeautifulSoup(r.content, 'html5lib') 
```

```python
print(soup.prettify()) 
```

```python
drugs = []
for div in soup.find_all('div', attrs ={'class': 'banned-para2'}):
    for li in div.find_all('li'):
        string = li.text
        string = string.replace('\xa0', ' ')
        drugs.append(string)
        #print(li.text)
```

```python
for div in soup.find_all('p', attrs ={'class': 'banned-para1'}):
    string = div.text
    string = string.replace('\xa0', ' ')
    drugs.append(string)
```

```python
import re
not_drug = []
words = []
for i in range(len(drugs)):
    try:
        word = drugs[i].split()
        k = int(word[0].strip("()"))
        words.append(k)
    except:
        word = drugs[i].split()
        not_drug.append(i)
```
